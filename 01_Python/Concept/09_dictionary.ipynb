# **Dictionary**

Dictionary in Python is a collection of keys values, used to store data values like a map, which, unlike other data types which hold only a single value as an element.

In some languages it is known as map or assosiative arrays.

dict = { 'name' : 'sharif', 'gender' : 'male' }

Characterstics:

- Mutable
- Indexing has no meaning
- keys can't be duplicated
- keys can't be mutable items

### **Create Dictionary**

````python
# empty dictionary
d = {}
print(d)

# 1D dictionary
d1 = { 'name' : 'last name' ,'gender' : 'male' }
print(d1)

# with mixed keys
d2 = {(1,2,3):1,'hello':'world'}
print(d2)

# 2D dictionary -> JSON
s = {
    'name':'first name',
     'college':'bit',
     'sem':4,
     'subjects':{
         'dsa':50,
         'maths':67,
         'english':34
     }
}
print(s)

# using sequence and dict function
d4 = dict([('name','first name'),('age',32),(3,3)])
print(d4)

# duplicate keys
d5 = {'name':'karim','name':'rahul'}
print(d5)

# mutable items as keys
d6 = {'name':'karim',(1,2,3):2}
print(d6)
````

### **Accessing items**

````python
my_dict = {'name': 'Jack', 'age': 26}
# []
print(my_dict['age'])
# get
print(my_dict.get('age'))

s = {
    'name':'first name',
     'college':'bit',
     'sem':4,
     'subjects':{
         'dsa':50,
         'maths':67,
         'english':34
     }
}
print(s['subjects']['maths'])
````

### **Adding key-value pair**

````python
d4 = dict([('name','first name'),('age',32),(3,3)])
print(d4)

d4['gender'] = 'male'
print(d4)

d4['weight'] = 72
print(d4)


s['subjects']['ds'] = 75
print(s)
````

### **Remove key-value pair**

````python
d = {'name': 'someone', 'age': 32, 3: 3, 'gender': 'male', 'weight': 72}
print(d)
# pop
d.pop(3)
print(d)

# popitem
d.popitem()
print(d)

# del
del d['name']
print(d)

# clear
d.clear()
print(d)

del s['subjects']['maths']
print(s)
````

### **Editing key-value pair**

````python
print(s)
s['subjects']['dsa'] = 80
print(s)
````

### **Dictionary Operations**

- Membership
- Iteration

````python
print(s)

print('name' in s)
````

````python
d = {'name':'karim','gender':'male','age':33}

for i in d:
  print(i,d[i])
````

### **Dictionary Functions**

````python
# len/sorted
len(d)
print(d)
sorted(d,reverse=True)
max(d)
````

````python
# items/keys/values
print(d)
print(d.items())
print(d.keys())
print(d.values())
````

````python
# update
d1 = {1:2,3:4,4:5}
d2 = {4:7,6:8}

d1.update(d2)
print(d1)
````

### **Dictionary Comprehension**

```bash
{ key: value for vars in iterable }
```

````python
# print 1st 10 numbers and their squares
{i:i**2 for i in range(1,11)}
````

````python
distances = {'dhaka':1000,'ctg':2000,'brahmanbaria':3000}
print(distances.items())
````

````python
# using existing dict
distances = {'dhaka':1000,'ctg':2000,'brahmanbaria':3000}
{key:value*0.62 for (key,value) in distances.items()}
````

````python
# using zip
days = ["Sunday", "Monday","Tuesday","Wednesday","Thursday","Friday","Saturday"]
temp_C = [30.5,32.6,31.8,33.4,29.8,30.2,29.9]

{i:j for (i,j) in zip(days,temp_C)}
````

````python
# using if condition
products = {'phone':10,'laptop':0,'charger':32,'tablet':0}

{key:value for (key,value) in products.items() if value>0}
````

````python
# Nested Comprehension
# print tables of number from 2 to 4
{i:{j:i*j for j in range(1,11)} for i in range(2,5)}
````
