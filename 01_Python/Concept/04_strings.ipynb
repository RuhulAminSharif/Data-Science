## **Creating String**

```python
# single line strings
s = 'a single line string'
print(s)
print()
t = "another single line string"
print(t)
print()
# multiline strings
s = '''multile line string using single qoutes
line 2
line 3'''
print(s)
print()
t  = """multile line string using double qoutes
line 2
line 3"""
print(t)
```

```python
# to insert single qoute in a string, cover it with double qoutes
str = "I'm a student"
print(str)
# or
str = 'I\'m a student'
print(str)
# vice versa also valid
str = 'I"m a student'
print(str)
```

## **Accessing Substrings from a String**

```python
# Indexing / Positive Indexing
# Accesing from left
# indexing start from 0
# range of index 0 to len - 1 ( len is the length of the string)
s = 'hello world'
print(s[0])
print(s[1])
```

```python
# Negative Indexing
# Accessing from right
# -1 means last charater of the strings or first character from the right
s = 'hello world'
print(s[-1])
print(s[-2])
```

```python
# Slicing
s = 'hello world'
# to extract characters starting from at index 0 to (5-1)
print(s[0:5])
# to extract characters from the 0th index to (5-1)th character
print(s[:5])
# to extract characters starting from at index 6 to the last
print(s[6:])
# to extract characters starting from at index 0 to the 5th index charater but jumping by 2character towards right
print(s[0:6:2])
# to extract characters starting from at index 6 to the 1th index charater but jumping by 2character towards left
print(s[6:0:-2])
# to reverse a string
print(s[::-1])
# to print from the 5th charater from right to the end of the string
print(s[-5:])
#
print(s[-1:-6:-1])
```

### **Capitalize/Title/Upper/Lower/Swapcase**

```python
s = 'hello world'

# will capitalize the first character of the sentence
print(s.capitalize())

# will capitalize the first character of each word of the sentence
print(s.title())

# will make all character in uppercase
print(s.upper())

# will make all character in lowercase
print(s.lower())

# upper -> lower, lower -> upper
print(s.swapcase())
```

```python
# to check wheter a substring exist or not
print(s.find('l'))
print(s.find('ls'))
```

```python
# to check wheter a substring exist or not
print(s.index('l'))
print(s.index('ls'))
```

### **Format**

```python
name = 'Sharif'
gender = 'male'
print('Hi my name is {} and I am a {}'.format(name,gender))
print('Hi my name is {0} and I am a {1}'.format(name,gender))
print('Hi my name is {1} and I am a {0}'.format(gender,name))
```

### **isalnum/isalpha/isdigit/isidentifier**

```python
s = 'hello world'
# islnum checks whether it's alphabet or number
print(s.isalnum())

# isalpha checks whether is's alphabet
print(s.isalpha())

# isdigit checks whether it's digit
print(s.isdigit())

# isidentifier checks for valid identifier
print(s.isidentifier())

s = 'helloworld'
print(s.isalnum())
print(s.isalpha())
print(s.isdigit())
print(s.isidentifier())
```

### **Split/Join**

```python
s = "My name is Ruhul Amin Sharif"

# split function will split the strings word by word and then convert into list
lst = s.split()  # by defalut, strings will be splitted using 'space'
print(lst)

# To split based on other charater
lst = s.split('a')
print(lst)

# to split baseed on string
lst = s.split('is')
print(lst)

```

```python
s = "My name is Ruhul Amin Sharif"
# split function will split the strings word by word and then convert into list
lst = s.split()  # by defalut, strings will be splitted using 'space'
print(lst)

retain = " ".join(lst)
print(retain)

retain = "-".join(lst)
print(retain)
```

### **Replace**

```python
# Count the frequency of a particular character in a provided string.
# Eg 'hello how are you' is the string, the frequency of h in this string is 2.
s = input()
term = input()
freq = 0
for i in s:
    if i == term:
        freq += 1
print(freq)
```

```python
# Write a program which can remove a particular character from a string.
s = input()
term = input()
res = ''
for i in s:
    if i != term:
        res += i
print(res)
```

```python
# Write a program that can check whether a given string is palindrome or not.
# abba
# using shortcut
s = input()
if s == s[::-1]:
    print('Palindrome')
else:
    print('Not Palindrome')
```

```python
# Write a program to count the number of words in a string without split()
s = input()
if s == '':
    print(0)
else:
    count = 1
    for i in s:
        if i == ' ':
            count += 1
    print(count)
```

```python
# Write a python program to convert a string to title case without using the title()
s = input()
res = ''
for i in range(len(s)):
    if i == 0 or s[i-1] == ' ':
        res += s[i].upper()
    else:
        res += s[i].lower()
print(res)
```

```python
# Write a program that can convert an integer to string.

number = int(input('enter the number'))

digits = '0123456789'
result = ''
while number != 0:
  result = digits[number % 10] + result
  number = number//10

print(result)
print(type(result))
```
