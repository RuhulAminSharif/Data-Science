## **How lists are stored in memory**

```python
L = [1, 2, 3]

print(id(L))
print(id(L[0]))
print(id(L[1]))
print(id(L[2]))
print(id(1))
print(id(2))
print(id(3))
```

## **Creating a List**

```python
# Empty
print([])
# 1D -> Homo
print([1, 2, 3, 4, 5])
# 2D
print([ 1, 2, 3, [4, 5] ])
# 3D
print([ [ [1, 2], [3, 4] ], [ [5, 6], [7, 8] ]])
# Hetrogenous
print([1, True, 5.6, 5+6j,'Hello'])
# Using Type conversion
print(list('hello'))
```

## **Accessing Items from a List**

```python
# Indexing
L = [[[1,2],[3,4]],[[5,6],[7,8]]]

# Indexing / Positive Indexing
# Accesing from left
# indexing start from 0
print(L[0][0][1])

# Negative Indexing
# Accessing from right
print(L[-1])

# Slicing
L = [1,2,3,4,5,6]
# to extract elements starting from at index 0 to (5-1)
print(L[0:5])
# to extract elements from the 0th index to (5-1)th character
print(L[:5])
# to extract elements starting from at index 6 to the last
print(L[6:])
# to extract elements starting from at index 0 to the 5th index charater but jumping by 2character towards right
print(L[0:6:2])
# to extract elements starting from at index 6 to the 1th index charater but jumping by 2character towards left
print(L[6:0:-2])
# to reverse a string
print(L[::-1])
# to print from the 5th charater from right to the end of the string
print(L[-5:])
#
print(L[-1:-6:-1])
```

## **Editing items in a List**

```python
L = [1,2,3,4,5]

# editing with indexing
L[0] = 4
print(L)

L[-1] = 500
print(L)

# editing with slicing
L[1:4] = [200,300,400]

print(L)
```

## **Deleting items from a List**



*   del
*   remove

```python
# del
L = [1,2,3,4,5]

# indexing
del L[-1]

# slicing
del L[1:3]
print(L)
```

```python
# pop
L = [1,2,3,4,5]

# pop with index
L.pop(0)
print(L)

# default pop
L.pop()
print(L)
```

## **List Comprehension**
List Comprehension provides a concise way of creating lists.

newlist = [expression for item in iterable if condition == True]

Advantages of List Comprehension

* More time-efficient and space-efficient than loops.
* Require fewer lines of code.
* Transforms iterative statement into a formula.

```python
# Add 1 to 10 numbers to a list
L = []

for i in range(1,11):
  L.append(i)

print(L)
```

```python
L = [i for i in range(1,11)]
print(L)
```

```python
# scalar multiplication on a vector
v = [2,3,4]
s = -3
# [-6,-9,-12]

x = []
for i in v:
    x.append(i * s )
print(x)

# using list comprehension
y = [s*i for i in v]
print(y)
```

```python
# Add squares
L = [1,2,3,4,5]

[i**2 for i in L]
```

```python
# Print all numbers divisible by 5 in the range of 1 to 50

[i for i in range(1,51) if i%5 == 0]
```

```python
# find languages which start with letter p
languages = ['java','python','php','c','javascript']

[language for language in languages if language.startswith('p')]
```

```python
# Nested if with List Comprehension
basket = ['apple','guava','cherry','banana']
my_fruits = ['apple','kiwi','grapes','banana']

# add new list from my_fruits and items if the fruit exists in basket and also starts with 'a'

[fruit for fruit in my_fruits if fruit in basket if fruit.startswith('a')]
```

```python
# Print a (3,3) matrix using list comprehension -> Nested List comprehension
[[i*j for i in range(1,4)] for j in range(1,4)]
```

```python
# cartesian products -> List comprehension on 2 lists together
L1 = [1,2,3,4]
L2 = [5,6,7,8]

[i*j for i in L1 for j in L2]
```

## **2 ways to traverse a list**

- itemwise
- indexwise

```python
# itemwise
L = [1,2,3,4]

for i in L:
  print(i)
```

```python
# indexwise
L = [1,2,3,4]

for i in range(0,len(L)):
  print(L[i])
```

## **Zip**

The zip() function returns a zip object, which is an iterator of tuples where the first item in each passed iterator is paired together, and then the second item in each passed iterator are paired together.

If the passed iterators have different lengths, the iterator with the least items decides the length of the new iterator.

```python
# Write a program to add items of 2 lists indexwise

L1 = [1, 2, 3, 4]
L2 = [-1,-2,-3,-4]

print(list(zip(L1,L2)))

print([i+j for i,j in zip(L1,L2)])
```
