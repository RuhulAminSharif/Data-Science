## **Operators in Python**

- Arithmetic Operators
- Relational Operators
- Logical Operators
- Bitwise Operators
- Assignment Operators
- Membership Operators

### **Arithmetic Operators**

```python
# addition
print( 5 + 6 )

# subtraction
print( 5 - 6)

# multiplication
print( 5 * 6 )

# division
print( 5 / 2 )

# integer division
print( 5 // 2 )

# modulus - remainder
print( 5 % 2 )

# power of
print( 5 ** 2 ) # 5 ^ 2
```

### **Relational Operators**

```python
# greater than
print( 4 > 5 )

# less than
print( 4 < 5 )

# greater than or equal to
print( 4 >= 4 )

# less than or equal to
print( 4 <= 4 )

# equal to
print( 4 == 4 )

# not equal to
print( 4 != 4 )
```

### **Logical Operators**

```python
# and operator
print(1 and 0)

# or operator
print(1 or 0)

# not operator
print(not 1)
```

### **Bitwise Operators**

```python

# bitwise and
print(2 & 3)

# bitwise or
print(2 | 3)

# bitwise xor
print(2 ^ 3)

# bitwise not
print(~3)

# bitwise rightshift
print(4 >> 2)

# bitwise leftshift
print(5 << 2)
```

### **Assignment Operators**

```python
# =
# a = 2

a = 2

# a = a % 2
a %= 2

# a++ ++a -> not valid in python

print(a)
```

### **Membership Operators**

```python
# in/not in

print('D' not in 'Dhaka')

print(1 in [2,3,4,5,6])
```

```python
# Program - Find the sum of a 3 digit number entered by the user

number = int(input('Enter a 3 digit number'))

# 345%10 -> 5
a = number%10

number = number//10

# 34%10 -> 4
b = number % 10

number = number//10
# 3 % 10 -> 3
c = number % 10

print(a + b + c)
```
