## **If-else in Python**

```python
# login program and indentation
# email -> name.company@gmail.com
# password -> 1234

email = input('enter email: ')
password = input('enter password: ')

if email == "name.company@gmail.com" and password == "1234":
    print("Welcome")
elif email == "name.company@gmail.com" and password != "1234":
    print("Incorrect Password")
    password = input('enter password again: ')
    if password == "1234":
        print("Welcome, finally!")
    else:
        print("You can't do it")
else:
    print("Not Correct")
```

```python
# min of 3 number

a = int(input('first num: '))
b = int(input('second num: '))
c = int(input('third num: '))

if a < b and a < c:
    print("Smallest Number: ", a )
elif b < c:
    print("Smallest Number: ", b )
else:
    print("Smallest Number: ", c )
```

```python
# menu driven calculator
fnum = int(input("Enter the 1st Number: "))
snum = int(input("Enter the 2nd Number: "))

op = input("""
Enter the Operation
1. For Addition: '+'
2. For Subtraction: '-'
3. For Multiplication: '*'
4. For Division: '/'
""")

if op == '+':
    print(fnum + snum)
elif op == '-':
    print(fnum - snum)
elif op == '*':
    print(fnum * snum)
elif op == '/':
    print(fnum / snum)
else:
    print("Not valid operator")
```

### **While Loop**

```python
# While loop example -> program to print the table
num = int(input("Enter the number: "))

i = 1
while i <= 10:
    print(num, '*', i, '=', num * i)
    i += 1
```

```python
# while loop with else

x = 1
while x < 3:
    print(x)
    x += 1
else:
    print("limit crossed")
```

```python
# Guessing game

import random
jackpot = random.randint(1, 100)

guess = int(input("Enter your guess: "))
counter = 1
while guess != jackpot:
    if guess < jackpot:
        print("Wrong! Guess Higher")
    else:
        print("Wrong! Guess Lower")
    guess = int(input("Enter your guess: "))
    counter += 1
else:
    print("Correct Guess")
    print("Attempts:", counter)
```

### Program - The current population of a town is 10000. The population of the town is increasing at the rate of 10% per year. You have to write a program to find out the population at the end of each of the last 10 years.

#### Explanation :



To calculate the population for each year with a 10% increase, you can use a simpler equation based on the previous year's population. Let's assume the population of the previous year is represented by variable x.

The equation can be written as:

> Current Year Population = x * 1.1

In this equation, the current year's population is equal to the previous year's population multiplied by 1.1, representing a 10% increase.

To find the population of the previous year (x), we can rearrange the equation as follows:

> x = Current Year Population / 1.1

Using this simplified equation, if you have the current year's population (e.g., 10,000), you can divide it by 1.1 to calculate the population of the previous year.

This equation allows you to calculate the population for each year, assuming you know the population of the current year and want to find the population of the previous year.

```python
curr_pop = 10000

for i in range(10,0,-1):
  print(i,curr_pop)
  curr_pop /= 1.1
```

### Sequence sum

1/1! + 2/2! + 3/3! + ...

```python
# code here
n = int(input("Enter N: "))

sum = 0
fact = 1
for i in range(1, n + 1 ):
    fact *= i
    sum += ( i / fact )
    print( i, "/", i, "!", end=" ", sep ="")
    if i != n:
        print(" +", end = " ")
print("=", sum)
```

# **Loop Control Statement**

- Break
- Continue
- Pass

```python
# break
for i in range(1, 10):
    if i == 5:
        break
    print(i, end=" ")
```

```python
# printing prime numbers in a range

lower = int(input('Enter lower range: '))
upper = int(input('Enter upper range: '))

for i in range( lower, upper+1):
    for j in range(2, i):
        if i % j == 0:
            break
    else:
        print(i, end = " ")
```

```python
# continue
for i in range(1, 10):
    if i == 5:
        continue
    print(i)
```

```python
# pass
for i in range(1, 10):
    pass
```
