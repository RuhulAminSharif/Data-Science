# **Python Output**

## **print function**

```python
# Python is a case sensitive language
print('Hello World')
print("Ruhul Amin Sharif")
print(7)
print(7.7)
print(True)
```


## **print function with separator (sep) parameter**

```python
print('Hello',1,4.5,True) # Default separator is ' ' (Space)
print('Hello',1,4.5,True,sep='/') # To change default separator
```

## **print function with end parameter**

```python
# Default end is newline
print("hello")
print("world")
```

```python
# To change the parameter of default end
print('hello',end='-')
print('world')
```

# **Literals**

## **Numeric Literals**

```python
a = 0b1010 #Binary Literals
b = 100 #Decimal Literal
c = 0o310 #Octal Literal
d = 0x12c #Hexadecimal Literal

#Float Literal
float_1 = 10.5
float_2 = 1.5e2 # 1.5 * 10^2
float_3 = 1.5e-3 # 1.5 * 10^-3

#Complex Literal
x = 3.14j

print(a, b, c, d)
print(float_1, float_2,float_3)
print(x, x.imag, x.real)
```

## **String Literals**

```python
string = 'This is Python'
strings = "This is Python"
char = "C"
multiline_str = """This is a multiline string with more than one line code."""
unicode = u"\U0001f600\U0001F606\U0001F923"
raw_str = r"raw \n string"

print(string)
print(strings)
print(char)
print(multiline_str)
print(unicode)
print(raw_str)
```
