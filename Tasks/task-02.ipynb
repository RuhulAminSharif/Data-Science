### **`Problem 1`** Write a program that will give you in hand monthly salary after deduction on CTC - HRA(10%), DA(5%), PF(3%) and taxes deduction as below:

> Salary(Lakhs) : Tax(%)

*   Below 5 : 0%
*   5-10 : 10%
*   10-20 : 20%
*   aboove 20 : 30%

````python
# Write code here
def calculate_tax(income):
    """
    Calculates tax based on the provided progressive slabs:
    - Below 5 Lakhs: 0%
    - 5-10 Lakhs: 10%
    - 10-20 Lakhs: 20%
    - Above 20 Lakhs: 30%
    """
    tax = 0

    # Slab 1: 0 - 5 Lakhs
    if income <= 500000:
        return 0

    # Slab 2: 5 Lakhs - 10 Lakhs
    # Tax the amount above 5L at 10%
    if income > 500000:
        taxable_amount = min(income, 1000000) - 500000
        tax += taxable_amount * 0.10

    # Slab 3: 10 Lakhs - 20 Lakhs
    # Tax the amount above 10L at 20%
    if income > 1000000:
        taxable_amount = min(income, 2000000) - 1000000
        tax += taxable_amount * 0.20

    # Slab 4: Above 20 Lakhs
    # Tax the amount above 20L at 30%
    if income > 2000000:
        taxable_amount = income - 2000000
        tax += taxable_amount * 0.30

    return tax

print("--- Monthly In-Hand Salary Calculator ---")
ctc_input = float(input("Enter your Annual CTC (in Lakhs): "))

# Convert Lakhs to absolute value for calculation
ctc_absolute = ctc_input * 100000

# Calculate Percentage Deductions
hra_deduction = ctc_absolute * 0.10  # 10%
da_deduction = ctc_absolute * 0.05   # 5%
pf_deduction = ctc_absolute * 0.03   # 3%

# Calculate Tax Deduction
tax_deduction = calculate_tax(ctc_absolute)

# Calculate Total Deductions
total_deductions = hra_deduction + da_deduction + pf_deduction + tax_deduction

# Calculate In-Hand Salary
annual_in_hand = ctc_absolute - total_deductions
monthly_in_hand = annual_in_hand / 12

# Output Results
print("\n--- Salary Breakdown ---")
print(f"Annual CTC:          ₹ {ctc_absolute:,.2f}")
print(f"HRA Deduction (10%): ₹ {hra_deduction:,.2f}")
print(f"DA Deduction (5%):   ₹ {da_deduction:,.2f}")
print(f"PF Deduction (3%):   ₹ {pf_deduction:,.2f}")
print(f"Tax Deduction:       ₹ {tax_deduction:,.2f}")
print("-" * 30)
print(f"Total Deductions:    ₹ {total_deductions:,.2f}")
print(f"Annual In-Hand:      ₹ {annual_in_hand:,.2f}")
print("-" * 30)
print(f"Monthly In-Hand:     ₹ {monthly_in_hand:,.2f}")
````

### **`Problem 7`** Reverse a given integer number.

Example:

`Input:`

```bash
76542
```

`Output:`

```bash
24567
```

````python
# Write code here
num = int(input("Enter the number: "))
rev = 0
print("Before reverse", num)
while num:
    digit = num % 10
    num //= 10
    rev = rev * 10 + digit
print("After reverse", rev)
````

### **`Problem 8`** Take a user input as integer N. Find out the sum from 1 to N. If any number if divisible by 5, then skip that number. And if the sum is greater than 300, don't need to calculate the sum further more. Print the final result.

**Example 1:**

`Input:`

```bash
30
```

`Output:`

```bash
276
```

````python
# Write code here
N = int(input("enter N: "))
tot = 0
for i in range(1, N+1):
    if i % 5 == 0:
        continue
    tot += i
    if tot > 300:
        tot -= i
        break
print(tot)
````

### **`Problem 11`** Write a program, which will find all such numbers between 1000 and 3000 (both included) such that each digit of the number is an even number. The numbers obtained should be printed in a space-separated sequence on a single line.



````python
# Write code here
for i in range(1000, 3001):
    flag = True
    curr = i
    while curr:
        d = curr % 10
        if d % 2 != 0:
            flag = False
            break
        curr //= 10
    if flag:
        print(i, end=",")
````

### **`Problem 12`** A robot moves in a plane starting from the original point (0,0). The robot can move toward UP, DOWN, LEFT and RIGHT with a given steps.
The trace of robot movement is shown as the following:
```
UP 5
DOWN 3
LEFT 3
RIGHT 2
!
```
> The numbers after the direction are steps.

> `!` means robot stop there.

**Please write a program to compute the distance from current position after a sequence of movement and original point.**

*If the distance is a float, then just print the nearest integer.*

Example:

`Input`:
```
UP 5
DOWN 3
LEFT 3
RIGHT 2
!
```
`Output`:
```
2
```

````python
# Write code here
x = 0
y = 0
while True:
    path = input("Enter the robot path: ")
    if path == "!":
        break
    dir = path.split()[0]
    steps = int(path.split()[1])
    if dir == "UP":
        y += steps
    elif dir == "DOWN":
        y -= steps
    elif dir == "RIGHT":
        x += steps
    else:
        x -= steps
print("distance: ", (x**2+y**2)**0.5)
````

### **`Problem 14`** Print all the Armstrong numbers in a given range.
Range will be provided by the user<br>
Armstrong number is a number that is equal to the sum of cubes of its digits. For example 0, 1, 153, 370, 371 and 407 are the Armstrong numbers.

````python
start_range = int(input("Enter the start of the range: "))
end_range = int(input("Enter the end of the range: "))

print(f"Armstrong numbers between {start_range} and {end_range} are:")

for num in range(start_range, end_range + 1):
    order = len(str(num))
    tot = 0
    temp_num = num
    while temp_num > 0:
        digit = temp_num % 10
        tot += digit ** order
        temp_num //= 10
    if num == tot:
        print(num)
````

### **`Problem 15`** Calculate the angle between the hour hand and minute hand.

Note: There can be two angles between hands; we need to print a minimum of two. Also, we need to print the floor of the final result angle. For example, if the final angle is 10.61, we need to print 10.

Input:<br>
H = 9 , M = 0<br>
Output:<br>
90<br>
Explanation:<br>
The minimum angle between hour and minute
hand when the time is 9 is 90 degress.

````python
H = int(input("Enter the hour (0-12): "))
M = int(input("Enter the minute (0-59): "))

# Validate input
if not (0 <= H <= 12 and 0 <= M <= 59):
    print("Invalid input. Hour must be between 0-12 and Minute between 0-59.")
else:
    H %= 12
    angle = (60 * H - 11 * M ) / 2
    angle = abs(angle)
    angle = min(angle, 360-angle)
    # Print the floor of the result
    print((angle))
````
