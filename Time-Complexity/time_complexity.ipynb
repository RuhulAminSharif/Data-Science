```python
number = int(input('enter the number'))

digits = '0123456789' 
result = ''
while number != 0:
  result = digits[number % 10] + result
  number = number // 10

print(result) # The time complexity -> O(log(n))
```

```python
L = [1,2,3,4]

sum = 0
for i in L:
  sum = sum + i

product = 1
for i in L:
  product = product*i

print(sum,product) # The time complexity -> O(n) for both
```

```python
L = [1,2,3,4,5]

for i in range(0,len(L)//2):
  other = len(L) - i -1
  temp = L[i]
  L[i] = L[other]
  L[other] = temp

print(L) # O(n)
```

```python
def fib(n):
  if n == 1 or n == 0:
    return 1
  else:
    return fib(n-1) + fib(n-2) 

# How the time complexity is measured here,
# Basically you have to count the number of function calls for a giver input
# Here for each function call two sub function calls are executing from n to 0 (roughly) 
# and for those 2 sub-calls next two sub-calls are running for each - TC - O(2 ^ n)
# meaning for each fn calls how many sub fn calls are running.   


print(fib(5))
```

```python
# How to imporve the fibonacchi recursion
def fib(n, d: dict):
  if n == 1 or n == 0:
    return 1
  if n in d:
    return d[n]
  else:
    val = fib(n-1, d) + fib(n-2, d)
    d[n] = val
    return val

d = {}
print(fib(5, d))
```
